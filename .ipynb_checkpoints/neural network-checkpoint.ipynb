```python
# Import all packages and libraries for NN

from keras.models import Sequential
from keras.layers import Dense
from sklearn.utils import shuffle
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


import keras
import pandas
import numpy
```

```python
# Data Processing
# Preparing Iris dataset for usage as a dataframe
# Encoding output variables
# Shuffle rows randomly for variety
# Spliting dataset to test and train dataframes. Test dataset will be 80% of entire dataframe.
# Parsing dataframe to features and labels (x_train and y_train)

# Constants
_training_split = 0.8

# Load in dataset
whole_df = pandas.read_csv('iris.csv')
whole_df = shuffle(whole_df)

# mapping = {'setosa': 1, 'versicolor': 2, 'virginica': 3}
# for keys in mapping.keys():
#     whole_df = whole_df.replace(keys, mapping[keys])

# Spliting dataframe to training and testing 
train_df = whole_df[:int(len(whole_df)*_training_split)]
test_df = whole_df[int(len(whole_df)*_training_split):]

# Turning training dataframe to x_train and encoded y_train
headers = list(whole_df.columns.values)
features = headers[:-1]
labels = headers[-1]

x_train = train_df[features].values
y_train = train_df[labels].values
input_dim = x_train.shape[1]

encoder = LabelEncoder()
encoder.fit(y_train)
encoded_train_Y = encoder.transform(y_train)
new_y_train = np_utils.to_categorical(encoded_train_Y)

# Turning testing dataframe to x_test and encoded y_test
x_test = test_df[features].values
y_test = test_df[labels].values

encoder = LabelEncoder()
encoder.fit(y_test)
encoded_test_Y = encoder.transform(y_test)
new_y_test = np_utils.to_categorical(encoded_test_Y)
```

```python
# Model Definition

model = Sequential()
model.add(Dense(8, input_dim=input_dim, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
# Model Compilation

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Model Training

model.fit(x_train, new_y_train, epochs=500)
```

```python
scores = model.evaluate(x_test, new_y_test, batch_size=128)
print("{}: {}%".format(model.metrics_names[1], scores[1]*100))
```
